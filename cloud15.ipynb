```
import numpy as np # linear algebra
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
```

```
#mpl.rcParams['figure.figsize']=(12.0,10.0)    #(6.0,4.0)
mpl.rcParams['font.size']=18                #10 
mpl.rcParams['savefig.dpi']=120             #72 
mpl.rcParams['figure.subplot.bottom']=.1 
```

```
stopwords = set(STOPWORDS)
```

```
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(['title, amt.pledged, blurb, by, country, currency, end.time, location, percentage.funded, state, story, type, Crowdfunding, has, become, one, of, the, main, sources, of, initial, capital, for, small, businesses, and, start-up, companies, that, are, looking, to, launch, their, first, products,']))
```

```
print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
fig.savefig("word1.png", dpi=900)
```
